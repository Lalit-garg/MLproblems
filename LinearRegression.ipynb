```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
intercept = 100
X, y, slope = make_regression(100, n_features=1, bias=intercept, coef=True, noise=15)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
class LinearRegression:
  def __init__(self,lr=0.1,iterations=100):
    self.lr=lr
    self.iterations=iterations

  def fit(self,X,y):
    self.X=X
    self.y=y
    
    self.intercept=0
    self.slope=0
    errors=[]
    for i in range(self.iterations):
      plt.scatter(self.X,y)
      # print(self.slope,self.intercept,end="\n")
      self.gradient()
      # print(self.slope,self.intercept,end="\n")
      plt.scatter(self.X,LinearRegression.function(self.X,self.slope,self.intercept))
      plt.show()
      plt.clf()
      
      errors.append(self.error())

    return errors

  def gradient(self):
    n_intercept,n_slope=self.gradient_dec()

    self.intercept=self.intercept-self.lr*n_intercept
    self.slope=self.slope-self.lr*n_slope

  
  def error(self):
    yh=LinearRegression.function(self.X,self.slope,self.intercept)
    diff=yh-self.y.reshape(-1,1)
    return np.sum(diff**2)
  
  def gradient_dec(self):
    yh=LinearRegression.function(self.X,self.slope,self.intercept)
    diff=yh-self.y.reshape(-1,1)
    n_intercept=np.sum(diff)/len(self.X)
    n_slope=np.sum(diff*self.X)/len(self.X)

    return n_intercept,n_slope

  def predict(self,X):
    return LinearRegression.function(X,self.slope,self.intercept).flatten()

  # def score(self,X,y):
  #   y_predicted=self.predict(X)
  #   return np.sum(abs(y_predicted-y)/(y))/len(y)

  @classmethod
  def function(cls,X,slope,intercept):
    return X*slope + intercept
```

```python
li=LinearRegression()
```

```python
error=li.fit(X_train,y_train)
```
